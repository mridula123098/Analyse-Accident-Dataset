```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
import seaborn as sns
```

```python
acc=pd.read_csv("Accident_Information.csv")
```

```python
acc.columns
```

Filtering only the required columns

```python
drop_cols=[ 'Accident_Index', '1st_Road_Class', '1st_Road_Number', '2nd_Road_Class',
       '2nd_Road_Number','Did_Police_Officer_Attend_Scene_of_Accident',
       'Junction_Control', 'Junction_Detail','Local_Authority_(District)', 'Local_Authority_(Highway)',
       'Location_Easting_OSGR', 'Location_Northing_OSGR',
       'LSOA_of_Accident_Location','Number_of_Vehicles', 'Pedestrian_Crossing-Human_Control',
       'Pedestrian_Crossing-Physical_Facilities', 'Police_Force',
       'Road_Surface_Conditions','Special_Conditions_at_Site','Urban_or_Rural_Area','InScotland',"Carriageway_Hazards","Year"]
acc.drop(drop_cols,axis=1,inplace=True)
```

```python
acc["Date"]=pd.to_datetime(acc["Date"])
```

```python
acc=acc[acc["Date"]>='2017-01-01']
```

```python
acc.reset_index(drop=True,inplace=True)
```

```python
acc["Month"]=acc["Date"].dt.month
```

```python
acc.head()
```

```python
sns.scatterplot(y="Latitude",x="Longitude",hue="Accident_Severity",palette="viridis",data=acc)
```

### Slight Accident Severity is mostly seen among the accidents in UK

```python
sns.countplot(data=acc,x="Accident_Severity",palette="Paired")
```

```python
sns.countplot(data=acc,x="Month",palette="rocket");
```

### In the year 2017, month of November has most of the accidents

```python
sns.countplot(data=acc,x="Day_of_Week",palette="hls").set_title("Frequency of accidents per day ");
```

```python
sns.countplot(data=acc,x="Speed_limit",palette='husl')
```

### A speed limit of 30kmph is in victim's vehicle

```python
def other_con(weather):
    if (weather in ['Unknown','Data missing or out of range'] ):
        return "Other"
    else: return weather
acc["Weather_Conditions"]=acc["Weather_Conditions"].apply(other_con)        
```

```python
sns.countplot(data=acc,x="Accident_Severity",hue="Weather_Conditions")
```

### The weather factor seems to have no effect on accidents occurance

```python
def others(light):
    if(light in ['Darkness - lighting unknown', 'Darkness - lights unlit',
       'Data missing or out of range'] ):
        return "Other"
    else: return light
    
acc['Light_Conditions']=acc["Light_Conditions"].apply(others)
```

```python
sns.scatterplot(y="Number_of_Casualties",x="Light_Conditions",hue="Accident_Severity",palette='rocket',data=acc)
```
